```python
def get_stochastic(df, n=15, m=5, t=3):
    # https://excelsior-cjh.tistory.com/111
    # 입력받은 값이 dataframe이라는 것을 정의해줌
    df = pd.DataFrame(df)
    # n일중 최고가
    ndays_high = df.High.rolling(window=n, min_periods=1).max()
    # n일중 최저가
    ndays_low = df.Low.rolling(window=n, min_periods=1).min()
 
    # Fast%K 계산
    kdj_k = ((df.Close - ndays_low) / (ndays_high - ndays_low))*100
    # Fast%D (=Slow%K) 계산
    kdj_d = kdj_k.ewm(span=m).mean()
    # Slow%D 계산
    kdj_j = kdj_d.ewm(span=t).mean()
 
    # dataframe에 컬럼 추가
    df = df.assign(kdj_k=kdj_k, kdj_d=kdj_d, kdj_j=kdj_j).dropna()
    
    return df
```

```python
def get_high_idx(df, term, next_n):
    highs = []
    idx = term  # 52주 신고가
    n = len(df)
    
    while idx < n-1:
        df_cut = df['Close'][idx-term:idx+1]
        if df_cut.max() == df_cut.iloc[-1]:
            highs.append(idx)
            idx += next_n
        idx += 1
        
    return np.array(highs)
```

```python
def get_low_idx(df, term, next_n):
    highs = []
    idx = term  # 52주 신고가
    n = len(df)
    
    while idx < n:
        df_cut = df['Close'][idx-term:idx]
        if df_cut.min() == df_cut.iloc[-1]:
            highs.append(idx)
            idx += next_n
        idx += 1
        
    return np.array(highs)
```

```python
def get_ratio(df, idxes, later):
    n = len(df)
#    idxes = idxes[idxes > 0]
    idxes_next = idxes + later
    idxes = idxes[idxes_next < n]
    idxes_buy = idxes + 1
    idxes_next = idxes_next[idxes_next < n]


    ratios = df.iloc[idxes_next]['Close'].values / df.iloc[idxes_buy]['Close'].values
    stochas = df.iloc[idxes]['kdj_j'].values
    
    #test
    
    return ratios, stochas
```

```python
result_high = []
result_low = []

stoch_high = []
stoch_low = []
```

```python
for code in tqdm(codes[:20]):
    df = sql.extract_df(code)
#    df = df[df['Date'] >= '2019']
    df = get_stochastic(df)
    highs = get_high_idx(df, 250, 20)
    lows = get_low_idx(df, 250, 20)
    
    highs_list = get_ratio(df, highs, 10)
    lows_list = get_ratio(df, lows, 10)
    result_high.extend(highs_list[0].tolist())
    result_low.extend(lows_list[0].tolist())
    stoch_high.extend(highs_list[1].tolist())
    stoch_low.extend(lows_list[1].tolist())
```
