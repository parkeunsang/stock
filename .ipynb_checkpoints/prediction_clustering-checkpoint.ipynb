```python
def normalize(arr):
    if(np.std(arr)==0):
        return(np.zeros(len(arr)))
    return (arr-np.mean(arr))/np.std(arr)
```

```python
def adjMM(arr):
    if(np.std(arr)==0):
        return(np.zeros(len(arr)))
    return(arr/(np.median(arr)+1))
```

```python
df = df[df.index>='2018']
dfOpen = dfOpen[dfOpen.index>='2018']
dfVolume=dfVolume[dfVolume.index>='2018']
```

```python
#version1(주가만)
def makeDf1(rd, n,p=0):
    df_ = df.iloc[rd:rd+n,:] #첫번째 원소는 두번쨰의 change 구하기위해, 마지막원소는 예측값
    temp = df.iloc[rd+n+p,:]/dfOpen.iloc[rd+n,:]
    temp.name = '0'
    df_ = df_.append(temp)
    df_ = df_.dropna(axis=1)
    df_ = df_.reset_index(drop=True).T
    df_ = df_[df_.iloc[:,-1]<100]    
    df_.iloc[:,:-1]=df_.iloc[:,:-1].apply(lambda x:normalize(x),axis=1)
    return(df_)
```

```python
#거래량
def makeDf2(rd, n, p=0):

    df_ = df.iloc[rd:rd+n,:] #첫번째 원소는 두번쨰의 change 구하기위해, 마지막원소는 예측값
    
    temp = df.iloc[rd+n+p,:]/dfOpen.iloc[rd+n,:]
    temp.name = '0'
    df_ = df_.append(temp)
    #df_ = df_.append(df.iloc[rd+n+p,:]/dfOpen.iloc[rd+n,:])

    df_ = df_.dropna(axis=1)
    df_ = df_.reset_index(drop=True).T
    df_ = df_[df_.iloc[:,-1]<100]

    df_.iloc[:,:-1]=df_.iloc[:,:-1].apply(lambda x:normalize(x),axis=1)

    dfV = dfVolume.iloc[rd:rd+n,:].reset_index(drop=True).T
    dfV=dfV.loc[df_[df_.iloc[:,-1]<100].index,]
    dfV = dfV.apply(lambda x:normalize(x),axis=1)
    return(df_,dfV)
```

```python
n=7
p=0
np.random.seed(1984)
randomIdx = np.random.randint(len(df)-n-p,size=20)
```

```python
#version1
priceDf = pd.DataFrame()
for rd in randomIdx:
    temp = makeDf1(rd,7)
    priceDf = priceDf.append(temp)
```

### clustering

```python
#version 1
#cluster의 개수 10개로
kmP = KMeans(n_clusters=10, random_state = 2020) # random seed
kmP.fit(priceDf.iloc[:,:-1])
```

```python
#test
for i in range(10):
    idx = np.where(kmP.labels_==i)[0]
    res = priceDf.iloc[idx,-1]
    print(res.mean())
```

```python
#version2
#volume이랑 같이 보자
priceDf = pd.DataFrame()
volumeDf = pd.DataFrame()

for rd in randomIdx:
    temp = makeDf2(rd,n,p)
    priceDf = priceDf.append(temp[0])
    volumeDf = volumeDf.append(temp[1])
```

```python
n1=10
n2= 10
kmP = KMeans(n_clusters=n1, random_state = 2020) # random seed
kmV = KMeans(n_clusters=n2, random_state = 2020) # random seed
```

```python
kmP.fit(priceDf.iloc[:,:-1])
kmV.fit(volumeDf)
```

```python
result=[]
for i in range(n1):
    for j in range(n2):
        idx = np.where((kmP.labels_==i)&(kmV.labels_==j))[0]
        res = priceDf.iloc[idx,-1]
        #print(res.mean(), i, ' ', j, len(res))
        result.append([res.mean(),len(res),i,j])
        
```

```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
df = pd.concat([volumeDf,priceDf],axis=1)
```

```python
df
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(10, )),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(200, activation='relu'),
   tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dropout(0.2),

  
  tf.keras.layers.Dense(1)
])
```

```python
optimizer = tf.keras.optimizers.Adagrad(lr=0.01)
```

```python
model.compile(optimizer=optimizer,
              loss='binary_crossentropy')
```

```python
xTrain = df.iloc[:,:-1]
yTrain = df.iloc[:,-1]
```

```python
model.fit(xTrain,yTrain-1, epochs=100,batch_size=100, verbose=1)
```

```python
pred = model.predict(xTrain)
pred = pred.reshape(len(pred))
```

### DNN - 0 or 1

```python
xTrain = df.iloc[:,:-1]
yTrain = df.iloc[:,-1]

yTrain = (yTrain>1).astype(int)
```

```python
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(10, )),
    
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(200, activation='relu'),
   tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(32, activation='relu'),
  tf.keras.layers.Dropout(0.2),

  
  tf.keras.layers.Dense(1, activation = 'sigmoid')
])

```

```python
optimizer = tf.keras.optimizers.Adagrad(lr=0.001)
```

```python
model.compile(optimizer=optimizer,
              loss='binary_crossentropy',
             metrics=['accuracy'])
```

```python
model.fit(xTrain,yTrain, epochs=10,batch_size=100, verbose=1)
```

```python
pred = model.predict(xTrain)
pred = pred.reshape(len(pred))
```
